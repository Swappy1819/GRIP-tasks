# **Visulation of Dataset**

```python
# Plotting the dataset using scatter plot
sns.set_style('darkgrid')
sns.scatterplot(y= data['Scores'], x= data['Hours'])
plt.title('MARKS VS STUDY-HOURS',size = 15)
#plt.ylable('Percentage of marks', size = 8)
#plt.xlable('Hours studied', size = 8)
plt.show()
```

# **TRAINING THE MODEL**

***1. Preparing the data***

```python
# Defining X and Y From the Data

x , y = data.iloc[:, :-1], data.iloc[:, 1]
```

***2. Splitting the data***

```python
#For splitting the data
from sklearn.model_selection import train_test_split
x_train , x_test, y_train , y_test = train_test_split(x, y , test_size = 0.1 , random_state = 0) 
```

# **LINEAR REGRESSION**

```python
from sklearn.linear_model import LinearRegression
Reg = LinearRegression()
Reg.fit(x_train , y_train)

print ("THE MODEL IS NOW TRAINED")
```

# **Visualization of the trained model**

```python
# plotting the test data
plt.scatter(x,y)
plt.title("Hours vs Scores")
plt.xlabel("Hours studies")
plt.ylabel("Percentage score")

# plotting the regression line
line = Reg.coef_*x + Reg.intercept_

plt.plot (x, line , color = 'blue')
plt.show()
```

# **Prediction Analysis**

```python
# For testing data in Hours
print(x_test)

# For Predicting the scores
y_pred = Reg.predict(x_test)

print("-----------------------------------")

y_pred = Reg.predict(x_test)

print(y_pred)
```

# **Actual vs Prediction**

```python
# Difference between actual and predicted values

dv = pd.DataFrame({'Actual' : y_test , 'Predicted' : y_pred})
dv
```

# **Graphical analysis of Actual vs Predication**

```python
# Plotting the bar graph
dv.plot(kind='bar',figsize=(10,5))
plt.title('Hours vs Percentage')
plt.xlabel('Number of hours studied')
plt.ylabel('Percentage scored')
plt.show()
```

# **Prediction of the score of students**

```python
# Setting the given value in the dataset for percentage prediction

hrs = 9.25
pdt = Reg.predict([[hrs]])

print("No. of study Hours = {}".format(hrs))
print("Predicted percentage = {}".format(pdt[0]))
```

Hence,

As we can see that the No. of study hours is ***9.25*** and the percentage score is ***93.67%***

# **Percentage for the average students**

```python
# For calculating the average score of the student we used metrics from
from sklearn import metrics 

#For Mean absolute error
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_pred)) 

# for Mean Squared error
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))

# for Root Mean squared error
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
